```python
import requests
from bs4 import BeautifulSoup
```

```python
URL = "https://csi.ucdavis.edu/chapters/"
```

```python
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")
results = soup.find()
print(results.prettify())
```

```python
entry_titles = soup.find_all( class_ = "entry-title" )

counter = 1;

for entry_title in entry_titles:
    
    links = entry_title.find_all("a")
    for link in links:
        link_url = link["href"]
        print(f"This is frat {counter}: {link_url}\n")
        counter+=1;
```
